```python
# Filtrar Lima Metropolitana
df_lima = df[df['departamento_hecho'] == 'LIMA METROPOLITANA'].copy()
print(f"Registros Lima: {len(df_lima):,}")

# Ver valores exactos de materia para filtrar patrimonio
print(df_lima['materia_hecho'].unique())
```

```python
cols_modelo = [
    # Temporales
    'fecha_hora_hecho',      # de acá se deriva hora, día semana, mes, turno
    'turno_hecho',           # ya está calculado (mañana/tarde/noche/madrugada)
    'año_hecho',
    'mes_hecho',
    'dia_hecho',
    # Geográficas
    'distrito_hecho',
    'provincia_hecho',
    'departamento_hecho',
    # Tipo de delito
    'modalidad_hecho',
    'subtipo_hecho',
    'tipo_hecho',
    # IDs para trazabilidad
    'id_dgc',
    'ID_DGC_03',
    # Coordenadas (útiles para visualización geoespacial)
    'lat_hecho',
    'long_hecho',
    'ubigeo_hecho_delito',
]

df_lima_modelo = df_lima[cols_modelo].copy()
print(df_lima_modelo.shape)
```

```python
df_patrimonio = df_lima[df_lima['tipo_hecho'] == 'PATRIMONIO (DELITO)']
print(df_patrimonio['subtipo_hecho'].value_counts())
```

```python
# Filtro: Lima + Patrimonio (robos y hurtos)
df_lima_robos = df_lima[
    df_lima['tipo_hecho'] == 'PATRIMONIO (DELITO)'
][cols_modelo].copy()

print(f"Registros finales: {len(df_lima_robos):,}")
print(df_lima_robos['modalidad_hecho'].value_counts().head(20))
```

```python
# Variables temporales derivadas
df_lima_robos['hora'] = df_lima_robos['fecha_hora_hecho'].dt.hour
df_lima_robos['dia_semana'] = df_lima_robos['fecha_hora_hecho'].dt.dayofweek  # 0=lunes
df_lima_robos['nombre_dia'] = df_lima_robos['fecha_hora_hecho'].dt.day_name()

# Indicador día laborable vs feriado (básico, sin feriados peruanos por ahora)
df_lima_robos['es_laborable'] = (df_lima_robos['dia_semana'] < 5).astype(int)

# Semana del mes
df_lima_robos['semana_mes'] = df_lima_robos['fecha_hora_hecho'].apply(
    lambda x: (x.day - 1) // 7 + 1
)

# Verificar
print(df_lima_robos[['fecha_hora_hecho','hora','dia_semana','es_laborable','semana_mes']].head())
```

```python
# Contar denuncias por distrito y mes
conteo = df_lima_robos.groupby(['distrito_hecho', 'año_hecho', 'mes_hecho']).size().reset_index(name='freq')

# Calcular percentiles globales
p33 = conteo['freq'].quantile(0.33)
p66 = conteo['freq'].quantile(0.66)
print(f"P33: {p33:.0f} denuncias/mes | P66: {p66:.0f} denuncias/mes")

# Etiquetar niveles de riesgo
def etiquetar_riesgo(freq):
    if freq <= p33:
        return 'BAJO'
    elif freq <= p66:
        return 'MEDIO'
    else:
        return 'ALTO'

conteo['nivel_riesgo'] = conteo['freq'].apply(etiquetar_riesgo)
print(conteo['nivel_riesgo'].value_counts())

# Unir el target al dataframe principal
df_lima_robos = df_lima_robos.merge(
    conteo[['distrito_hecho', 'año_hecho', 'mes_hecho', 'nivel_riesgo', 'freq']],
    on=['distrito_hecho', 'año_hecho', 'mes_hecho'],
    how='left'
)

print(df_lima_robos['nivel_riesgo'].value_counts())
```

```python
# 1. Definir subtipos válidos (solo robos y hurtos)
subtipos_validos = [
    'HURTO',
    'ROBO',
    'HURTO DE VEHICULO',
    'HURTO DE CELULAR',
    'HURTO DE ACCESORIOS Y AUTOPARTES DE VEHICULOS',
    'HURTO AGRAVADO EN CASA HABITADA',
    'ASALTO Y ROBO DE VEHICULOS',
    'TENTATIVA DE ROBO',
    'TENTATIVA DE HURTO',
]

df_lima_robos = df_lima_robos[
    df_lima_robos['subtipo_hecho'].isin(subtipos_validos)
].copy()

print(f"Registros después de filtrar subtipos: {len(df_lima_robos):,}")
print(df_lima_robos['subtipo_hecho'].value_counts())
```

```python
# 2. Corregir el target: convertir distrito a string antes del groupby
df_lima_robos['distrito_hecho'] = df_lima_robos['distrito_hecho'].astype(str)

conteo = df_lima_robos.groupby(
    ['distrito_hecho', 'año_hecho', 'mes_hecho'],
    observed=True   # silencia el FutureWarning y evita combinaciones vacías
).size().reset_index(name='freq')

p33 = conteo['freq'].quantile(0.33)
p66 = conteo['freq'].quantile(0.66)
print(f"P33: {p33:.0f} | P66: {p66:.0f}")
print(f"Rango de frecuencias: {conteo['freq'].min()} a {conteo['freq'].max()}")
print(conteo['freq'].describe())
```

```python
# 3. Etiquetar y hacer merge
def etiquetar_riesgo(freq):
    if freq <= p33:
        return 'BAJO'
    elif freq <= p66:
        return 'MEDIO'
    else:
        return 'ALTO'

conteo['nivel_riesgo'] = conteo['freq'].apply(etiquetar_riesgo)

print("\nDistribución del target:")
print(conteo['nivel_riesgo'].value_counts())
print(conteo['nivel_riesgo'].value_counts(normalize=True).round(3))

df_lima_robos = df_lima_robos.merge(
    conteo[['distrito_hecho', 'año_hecho', 'mes_hecho', 'nivel_riesgo', 'freq']],
    on=['distrito_hecho', 'año_hecho', 'mes_hecho'],
    how='left'
)

print(f"\nDistribución final en df:")
print(df_lima_robos['nivel_riesgo'].value_counts())
```

```python
print(df_lima_robos.groupby(['año_hecho','mes_hecho']).size().sort_index())
print(f"\nMeses únicos disponibles: {df_lima_robos['mes_hecho'].nunique()}")
print(f"Distritos únicos: {df_lima_robos['distrito_hecho'].nunique()}")
print(f"\nDistribución por turno:")
print(df_lima_robos['turno_hecho'].value_counts())
```

```python
# ============================================================
# SPLIT CRONOLÓGICO (resuelve corrección del profesor)
# Entrenamiento: ene 2025 - dic 2025 (12 meses)
# Test: ene 2026 - may 2026 (5 meses, datos "futuros")
# ============================================================

df_train = df_lima_robos[df_lima_robos['año_hecho'] == 2025].copy()
df_test  = df_lima_robos[df_lima_robos['año_hecho'] == 2026].copy()

print(f"Train: {len(df_train):,} registros ({df_train['mes_hecho'].nunique()} meses)")
print(f"Test:  {len(df_test):,} registros ({df_test['mes_hecho'].nunique()} meses)")
```

```python
# ============================================================
# TARGET: nivel de riesgo por distrito + turno
# Construido SOLO con datos de entrenamiento (sin fuga)
# ============================================================

conteo_train = df_train.groupby(
    ['distrito_hecho', 'turno_hecho'],
    observed=True
).size().reset_index(name='freq')

# Percentiles calculados solo sobre train
p33 = conteo_train['freq'].quantile(0.33)
p66 = conteo_train['freq'].quantile(0.66)
print(f"P33: {p33:.0f} | P66: {p66:.0f}")
print(conteo_train['freq'].describe())

def etiquetar_riesgo(freq):
    if freq <= p33:   return 'BAJO'
    elif freq <= p66: return 'MEDIO'
    else:             return 'ALTO'

conteo_train['nivel_riesgo'] = conteo_train['freq'].apply(etiquetar_riesgo)
print("\nDistribución del target (sobre combinaciones distrito×turno):")
print(conteo_train['nivel_riesgo'].value_counts())
print(conteo_train['nivel_riesgo'].value_counts(normalize=True).round(3))
```

```python
# Eliminar columnas duplicadas del merge anterior
cols_a_eliminar = ['nivel_riesgo', 'freq']
df_train = df_train.drop(columns=[c for c in cols_a_eliminar if c in df_train.columns])
df_test  = df_test.drop(columns=[c for c in cols_a_eliminar if c in df_test.columns])

# Ahora sí hacer el merge
target_map = conteo_train[['distrito_hecho', 'turno_hecho', 'nivel_riesgo', 'freq']]

df_train = df_train.merge(target_map, on=['distrito_hecho', 'turno_hecho'], how='left')
df_test  = df_test.merge(target_map, on=['distrito_hecho', 'turno_hecho'], how='left')

# Verificar
print(f"Nulos en target (test): {df_test['nivel_riesgo'].isna().sum()}")
print(f"\nDistribución train:\n{df_train['nivel_riesgo'].value_counts()}")
print(f"\nDistribución test:\n{df_test['nivel_riesgo'].value_counts()}")
```

```python
# Eliminar el registro sin target en test
df_test = df_test.dropna(subset=['nivel_riesgo'])
print(f"Test final: {len(df_test):,} registros")

# Unir train y test en un solo CSV con columna que identifica el split
df_train['split'] = 'train'
df_test['split']  = 'test'
df_final = pd.concat([df_train, df_test], ignore_index=True)

print(f"Dataset final: {df_final.shape}")
print(f"Columnas: {df_final.columns.tolist()}")
```

```python
# Guardar como CSV
ruta_csv = '/content/drive/MyDrive/Colab Notebooks/lima_robos_limpio.csv'
df_final.to_csv(ruta_csv, index=False, encoding='utf-8-sig')

import os
tam = os.path.getsize(ruta_csv)
print(f"Guardado en: {ruta_csv}")
print(f"Tamaño: {tam / (1024**2):.1f} MB")
print(f"\nResumen final:")
print(f"  Total registros: {len(df_final):,}")
print(f"  Train: {len(df_train):,} | Test: {len(df_test):,}")
print(f"  Columnas: {df_final.shape[1]}")
print(f"\nDistribución target completa:")
print(df_final.groupby(['split','nivel_riesgo']).size().unstack())
```
