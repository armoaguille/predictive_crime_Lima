```python
# Split cronológico
df_train = df[df['año_hecho'] == 2025].copy()
df_test  = df[df['año_hecho'] == 2026].copy()

print(f"Train: {len(df_train):,} registros")
print(f"Test:  {len(df_test):,} registros")
```

```python
# Calcular frecuencias SOLO con train
conteo_train = df_train.groupby(
    ['distrito_hecho', 'turno_hecho'],
    observed=True
).size().reset_index(name='freq')

p33 = conteo_train['freq'].quantile(0.33)
p66 = conteo_train['freq'].quantile(0.66)
print(f"P33: {p33:.0f} | P66: {p66:.0f}")

def etiquetar_riesgo(freq):
    if freq <= p33:   return 'BAJO'
    elif freq <= p66: return 'MEDIO'
    else:             return 'ALTO'

conteo_train['nivel_riesgo'] = conteo_train['freq'].apply(etiquetar_riesgo)

print("\nDistribución del target (combinaciones distrito×turno):")
print(conteo_train['nivel_riesgo'].value_counts())
```

```python
# Aplicar el target a train y test (definido solo con train)
target_map = conteo_train[['distrito_hecho', 'turno_hecho', 'nivel_riesgo']]

df_train = df_train.merge(target_map, on=['distrito_hecho', 'turno_hecho'], how='left')
df_test  = df_test.merge(target_map, on=['distrito_hecho', 'turno_hecho'], how='left')

print(f"Nulos en target (train): {df_train['nivel_riesgo'].isna().sum()}")
print(f"Nulos en target (test):  {df_test['nivel_riesgo'].isna().sum()}")

print(f"\nDistribución train (registros):")
print(df_train['nivel_riesgo'].value_counts())
print(f"\nDistribución test (registros):")
print(df_test['nivel_riesgo'].value_counts())
```

```python
# Ver qué registro de test no tiene target
nulo_test = df_test[df_test['nivel_riesgo'].isna()]
print(nulo_test[['distrito_hecho', 'turno_hecho']])
```

```python
df_test = df_test.dropna(subset=['nivel_riesgo'])
print(f"Test final: {len(df_test):,} registros")
```

```python
print("Proporción de clases en train (registros):")
print(df_train['nivel_riesgo'].value_counts(normalize=True).round(3))
```

```python
orden_riesgo = {'BAJO': 0, 'MEDIO': 1, 'ALTO': 2}

df_train['nivel_riesgo_cod'] = df_train['nivel_riesgo'].map(orden_riesgo)
df_test['nivel_riesgo_cod']  = df_test['nivel_riesgo'].map(orden_riesgo)

print(df_train[['nivel_riesgo','nivel_riesgo_cod']].drop_duplicates())
```

```python
from sklearn.preprocessing import OneHotEncoder

cols_onehot = ['distrito_hecho', 'turno_hecho', 'subtipo_hecho']

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder.fit(df_train[cols_onehot])

# Transformar train
train_encoded = encoder.transform(df_train[cols_onehot])
train_encoded_df = pd.DataFrame(
    train_encoded,
    columns=encoder.get_feature_names_out(cols_onehot),
    index=df_train.index
)

# Transformar test
test_encoded = encoder.transform(df_test[cols_onehot])
test_encoded_df = pd.DataFrame(
    test_encoded,
    columns=encoder.get_feature_names_out(cols_onehot),
    index=df_test.index
)

print("Columnas generadas:", train_encoded_df.shape[1])
print(train_encoded_df.columns.tolist()[:10], "...")
```

```python
# Variables numéricas que se mantienen tal cual
cols_numericas = ['año_hecho', 'mes_hecho', 'dia_hecho', 'hora', 'dia_semana', 'lat_hecho', 'long_hecho']

df_train_final = pd.concat([
    df_train[cols_numericas].reset_index(drop=True),
    train_encoded_df.reset_index(drop=True),
    df_train[['nivel_riesgo_cod']].reset_index(drop=True)
], axis=1)

df_test_final = pd.concat([
    df_test[cols_numericas].reset_index(drop=True),
    test_encoded_df.reset_index(drop=True),
    df_test[['nivel_riesgo_cod']].reset_index(drop=True)
], axis=1)

print("Train final:", df_train_final.shape)
print("Test final:", df_test_final.shape)
df_train_final.head()
```

```python
from sklearn.preprocessing import StandardScaler

cols_escalar = ['año_hecho', 'mes_hecho', 'dia_hecho', 'hora', 'dia_semana', 'lat_hecho', 'long_hecho']

scaler = StandardScaler()
scaler.fit(df_train_final[cols_escalar])  # fit SOLO con train

# Crear copias escaladas (para Regresión Logística / KNN)
df_train_scaled = df_train_final.copy()
df_test_scaled  = df_test_final.copy()

df_train_scaled[cols_escalar] = scaler.transform(df_train_final[cols_escalar])
df_test_scaled[cols_escalar]  = scaler.transform(df_test_final[cols_escalar])

print(df_train_scaled[cols_escalar].describe())
```
